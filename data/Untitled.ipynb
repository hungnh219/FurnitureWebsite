```python
Data = pd.ExcelFile('Data.xlsx')
Table = Data.sheet_names
```

```python
def Get_print_value(Data_in, Name):
    It = Data_in.parse(Name)
    print(f"INSERT INTO {Name}")
    print(f"VALUE")
    temp = ""
    for id, row in It.iterrows():
        temp += f"\t("
        for item in row:
            temp += f"'{item}',"
        temp = temp[:-1]
        temp += f"),\n"
    temp = temp[:-2]
    temp += f";\n"
    print(temp)
```

```python
for item in Table:
    Get_print_value(Data, item)
```
